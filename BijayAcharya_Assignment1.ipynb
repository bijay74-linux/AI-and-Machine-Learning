```python
'''6. Adding HDI Category Column:
• Create a new column HDI Category that classifies each country into one of the four official
Human Development Index groups. The classification should be based on the HDI value for the
year 2022. Use the following categories and thresholds defined by the United Nations Development
Programme (UNDP):
HDI Category HDI Range (hdi)
Low < 0.550
Medium 0.550 – 0.699
High 0.700 – 0.799
Very ≥ 0.800
After creating this new column:
• verify that all countries are classified correctly,
• ensure the updated dataframe includes the new category column.
• Save the final dataframe as HDI category added.csv and include this file in your final
submission.
'''
```

```python
def categorize_hdi(value):
    if value < 0.550:
        return 'Low'
    elif value <= 0.699:
        return 'Medium'
    elif value <= 0.799:
        return 'High'
    else:
        return 'Very High'

df22['HDI_Category'] = df22['hdi'].apply(categorize_hdi)
df22[['country', 'hdi', 'HDI_Category']].head(10)
df22['HDI_Category'].value_counts()
df22.to_csv('HDI_category_added.csv', index=False)
```

```python
'''Outlier Detection:
• Detect outliers in HDI and GNI per Capita using the 1.5 × IQR rule.'''

sa_2022 = df_south_asia[df_south_asia['year'] == 2022]
# Detecting HDI outliers using IQR
q1_hdi = sa_2022['hdi'].quantile(0.25)
q3_hdi = sa_2022['hdi'].quantile(0.75)
iqr_hdi = q3_hdi - q1_hdi

hdi_outliers = sa_2022[
    (sa_2022['hdi'] < q1_hdi - 1.5*iqr_hdi) |
    (sa_2022['hdi'] > q3_hdi + 1.5*iqr_hdi)
]
# Detecting GNI per capita outliers using IQR
q1_gni = sa_2022['gross_inc_percap'].quantile(0.25)
q3_gni = sa_2022['gross_inc_percap'].quantile(0.75)
iqr_gni = q3_gni - q1_gni

gni_outliers = sa_2022[
    (sa_2022['gross_inc_percap'] < q1_gni - 1.5*iqr_gni) |
    (sa_2022['gross_inc_percap'] > q3_gni + 1.5*iqr_gni)
]
print("HDI Outliers:")
print(hdi_outliers[['country', 'hdi']])

print("\nGNI per Capita Outliers:")
print(gni_outliers[['country', 'gross_inc_percap']])
```

```python
'''Create a scatter plot of GNI per Capita vs HDI, highlighting the outliers in a different color.'''
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))

sns.scatterplot(
    data=sa_2022,
    x='gross_inc_percap',
    y='hdi',
    s=100,
    label='Normal'
)
sns.scatterplot(
    data=hdi_outliers,
    x='gross_inc_percap',
    y='hdi',
    s=100,
    label='HDI Outlier'
)
sns.scatterplot(
    data=gni_outliers,
    x='gross_inc_percap',
    y='hdi',
    s=100,
    label='GNI Outlier'
)
plt.xlabel('GNI per Capita')
plt.ylabel('Human Development Index (HDI)')
plt.title('GNI vs HDI (South Asia, 2022)')

plt.legend()
plt.show()
```

```python
'''6. Correlation Analysis:
• For each region, compute correlations of HDI with:
– Gender Development Index
– Life Expectancy Index
• Create scatter plots with trendlines for each correlation.
• Interpret the strength and direction of these relationships.'''

import numpy as np

def pearson_corr(x, y):
    x_mean = np.mean(x)
    y_mean = np.mean(y)
    numerator = np.sum((x - x_mean) * (y - y_mean))
    denominator = np.sqrt(np.sum((x - x_mean)**2) * np.sum((y - y_mean)**2))
    return numerator / denominator

print(" -- South Asia Correlations --")
for metric in metrics:
    r = pearson_corr(df_south_asia['hdi'], df_south_asia[metric])
    print(f"{metric} vs HDI: {r:.4f}")

print("\n -- Middle East Correlations --")
for metric in metrics:
    r = pearson_corr(df_middle_east['hdi'], df_middle_east[metric])
    print(f"{metric} vs HDI: {r:.4f}")
```

```python
'''Outlier Detection:
• Detect outliers in HDI {"hdi"} and GNI per Capita {"gross inc percap"} for each region using
the 1.5 × IQR rule.
• Create scatter plots highlighting outliers in a different color.
• Discuss the significance of these outliers.'''

import pandas as pd
import matplotlib.pyplot as plt

df_sa = pd.read_csv('HDI_SouthAsia_2020_2022.csv')
df_me = pd.read_csv('HDI_MiddleEast_2020_2022.csv')

for col in ['hdi', 'gross_inc_percap']:
    df_sa[col] = pd.to_numeric(df_sa[col], errors='coerce')
    df_me[col] = pd.to_numeric(df_me[col], errors='coerce')

def detect_outliers(df, column):
    q1 = df[column].quantile(0.25)
    q3 = df[column].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    return df[(df[column] < lower_bound) | (df[column] > upper_bound)]

sa_hdi_outliers = detect_outliers(df_sa, 'hdi')
sa_gni_outliers = detect_outliers(df_sa, 'gross_inc_percap')

me_hdi_outliers = detect_outliers(df_me, 'hdi')
me_gni_outliers = detect_outliers(df_me, 'gross_inc_percap')

print("South Asia HDI Outliers:\n", sa_hdi_outliers[['country', 'hdi']])
print("South Asia GNI Outliers:\n", sa_gni_outliers[['country', 'gross_inc_percap']])

print("\nMiddle East HDI Outliers:\n", me_hdi_outliers[['country', 'hdi']])
print("Middle East GNI Outliers:\n", me_gni_outliers[['country', 'gross_inc_percap']])
```

```python
import matplotlib.pyplot as plt

def plot_outliers(df, x_col, y_col, title):
    plt.figure(figsize=(6,4))

    # Calculating IQR and bounds
    q1 = df[y_col].quantile(0.25)
    q3 = df[y_col].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr

    # Separating normal data and outliers
    outliers = df[(df[y_col] < lower_bound) | (df[y_col] > upper_bound)]
    normal = df[(df[y_col] >= lower_bound) & (df[y_col] <= upper_bound)]

    plt.scatter(normal[x_col], normal[y_col], color='blue', label='Normal')
    plt.scatter(outliers[x_col], outliers[y_col], color='red', label='Outlier')

    plt.title(title)
    plt.xlabel(x_col)
    plt.ylabel(y_col)
    plt.legend()
    plt.grid(True)
    plt.show()

plot_outliers(df_sa, 'country', 'hdi', 'South Asia: HDI Outliers')
plot_outliers(df_sa, 'country', 'gross_inc_percap', 'South Asia: GNI per Capita Outliers')

plot_outliers(df_me, 'country', 'hdi', 'Middle East: HDI Outliers')
plot_outliers(df_me, 'country', 'gross_inc_percap', 'Middle East: GNI per Capita Outliers')
```

```python
""" In South Asia, there were no HDI or GNI outliers, suggesting that development levels are fairly consistent across the region. In the Middle East, however, Yemen stands out as a major exception, highlighting the severe development challenges it faces. Outliers are important because they reveal extreme cases and disparities within a region, helping policymakers focus on targeted interventions for countries that fall far below or above the regional average. """
```
